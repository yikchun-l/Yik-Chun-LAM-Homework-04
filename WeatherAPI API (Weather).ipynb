## 7) What's the weather looking like for the next 24+ hours in Miami, Florida?

I'd like to know the temperature for every hour, and if it's going to have cloud cover of more than 50% say "{temperature} and cloudy" instead of just the temperature. 

- *Tip: You'll only need one day of forecast*

```python
fl_url = "http://api.weatherapi.com/v1/forecast.json?key=99f080640936476091a124107221406&q=Miami&days=1&aqi=no&alerts=no"
fl_request = requests.get(fl_url)
fl_data = fl_request.json()
```

```python
for data in fl_data['forecast']['forecastday']:
    for details in data['hour']:
        hourly_time = details['time']
        hourly_temp = details['temp_c']
        hourly_cloud = details['cloud']
        if hourly_cloud > 50:
            print(f"{hourly_time} {hourly_temp} degrees C and cloudy")
        else:
            print(f"{hourly_time} {hourly_temp} degrees C")
```

## 8) For the next 24-ish hours in Miami, what percent of the time is the temperature above 85 degrees?

- *Tip: You might want to read up on [looping patterns](http://jonathansoma.com/lede/foundations-2017/classes/data%20structures/looping-patterns/)*

```python
over_85 = 0

for data in fl_data['forecast']['forecastday']:
    for details in data['hour']:
        hourly_time = details['time']
        hourly_temp = details['temp_f']
        hourly_cloud = details['cloud']
        if hourly_temp > 85:
            over_85 = over_85 + 1
            
percent_over_85 = (over_85 / len(hourly_time) * 100)
print("In the next 24 hours, Miami is over 85 degrees F", percent_over_85, "% of the time")
```
